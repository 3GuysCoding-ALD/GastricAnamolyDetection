## Creating test set

<b>Considering a random 80% and 20% spit for train , test </b>

<li>Creating a the sub directories</li>
<li>Moving the pictures with 80% to 20% random split</li>

```python
train_ratio = 0.8
test_ratio = 1-train_ratio
```

```python
for each in categories:
    if not os.path.exists(test_dir+'\\'+each):
        os.mkdir(test_dir+'\\'+each)
    if not os.path.exists(train_dir+'\\'+each):
        os.mkdir(train_dir+'\\'+each)
```

```python
for each in categories:
    src_dir = data_dir+'\\'+each
    lis_name_img = os.listdir(src_dir)
    train_split , test_split=train_test_split(lis_name_img, 
                                              test_size=0.2, 
                                              random_state=seed,
                                              shuffle = True)
    
    #print(len(train_split))
    for img in train_split:
        destination = train_dir+'\\'+each+'\\'+img
        source = data_dir+'\\'+each+'\\'+img
        
        if not os.path.exists(destination):
            copyfile(source, destination)
        
    for img in test_split:
        destination = test_dir+'\\'+each+'\\'+img
        source = data_dir+'\\'+each+'\\'+img
        
        if not os.path.exists(destination):
            copyfile(source, destination)
    
    #break
```
