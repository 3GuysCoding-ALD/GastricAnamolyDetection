```python
Base_Model = tf.keras.applications.InceptionV3(input_tensor=tf.keras.Input(shape=input_size),
                      weights='imagenet',
                      include_top=False)

conn_layer = Base_Model.get_layer('mixed7')
conn_output = conn_layer.output
x = conn_output 
x = Conv2D(512,(3,3),activation='relu')(x)
x = Conv2D(512,(3,3),activation='relu')(x)
x = BatchNormalization()(x)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = Dense(512,activation='relu')(x)
x = Dropout(0.1)(x)
x = Dense(128,activation='relu',name='feature_out')(x)
predictions = Dense(8,activation='softmax')(x)

customInception = tf.keras.Model(inputs=Base_Model.input, outputs=predictions)
```

```python
class Ensemble(tf.keras.Model):
    def __init__(self, customDensenet,customInception):
        super(Ensemble, self).__init__()
        self.DenseNet = tf.keras.Model(customDensenet.input,customDensenet.get_layer('feature_out').output)
        self.InceptionNet = tf.keras.Model(customInception.input,customInception.get_layer('feature_out').output)

        for layer in self.DenseNet.layers:
            layer.trainable = False

        for layer in self.InceptionNet.layers:
            layer.trainable = False

        clasification_model = tf.keras.Sequential([
            tf.keras.layers.Dense(256,activation='relu',name='classification_model_1'),
            tf.keras.layers.Dense(64,activation='relu',name='classification_model_2'),
            tf.keras.layers.Dense(8,activation='softmax',name='classification_model_3')
        ])

        self.concat = tf.keras.layers.Concatenate()
        self.clasificationModel = clasification_model

    def call(self, inputs):
        inception_out = self.InceptionNet(inputs)
        densenet_out = self.DenseNet(inputs)
        
        x = self.concat([inception_out,densenet_out])
        x = self.clasificationModel(x)
        return x
```

```python
ensemble = Ensemble(customDensenet,customInception)
```

```python
ensemble.compile(optimizer=tf.keras.optimizers.Adamax(lr=1e-2), 
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=[tf.keras.metrics.CategoricalAccuracy(name='accuracy'),
                       ]
              )
```
